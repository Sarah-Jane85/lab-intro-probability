**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

Binominal challenge

```python
from scipy.stats import binom
#tickets sold
n = 460
#chance of catching the flight
p = 0.97
seats = 450
#P( or fewer passengers show up)
probability = binom.cdf(seats, n, p)

print(f"Probability that all passengers get a seat: {probability:.4f}")
print(f"That's about {probability*100:.2f}%")
```

**Challenge 2**

#### Ironhack Call Center 

Suppose a customer service representative at a call center is handling customer complaints. Consider the following:
- The probability of successfully resolving a customer complaint on the first attempt is 0.3. 


What is the probability that the representative needs to make at least three attempts before successfully resolving a customer complaint?

Geometric distribution problem

```python
from scipy.stats import geom
# Parameters
p = 0.3  # probability of resolving on first attempt

# P(needs AT LEAST 3 attempts) = P(first success happens on attempt 3 or later)
probability = geom.sf(2, p)  # sf(2) means "more than 2 attempts", sf= survival function

print(f"Probability of needing at least 3 attempts: {probability:.4f}")
print(f"That's about {probability*100:.2f}%")
```

```python
# P(resolves on attempt 1 or 2)
probability_1_or_2 = geom.cdf(2, p) #cdf = Cumulative distribution function, For dummies: "What is the probability of getting X or LESS?"
print(f"Probability of resolving within 2 attempts: {probability_1_or_2*100:.2f}%")
```

**Challenge 3**

#### Ironhack Website

Consider a scenario related to Ironhack website traffic. Where:
- our website takes on average 500 visits per hour.
- the website's server is designed to handle up to 550 vists per hour.


What is the probability of the website server being overwhelmed?

Poisson distribution

```python
from scipy.stats import poisson
# Parameters
mu = 500     # average visits per hour #mu is greek letter for M (weird u)
capacity = 550  # server capacity

# P(more than 550 visits) = server overwhelmed
probability = poisson.sf(capacity, mu)

print(f"Probability of server being overwhelmed: {probability:.4f}")
print(f"That's about {probability*100:.2f}%")
```

What is the probability of being overwhelmed at some point during a day? (consider 24hours)

```python
# Probability of NOT being overwhelmed in a single hour
p_safe_one_hour = 1 - probability  # from previous result

# Probability of NOT being overwhelmed in ALL 24 hours
p_safe_24_hours = p_safe_one_hour ** 24

# Probability of being overwhelmed AT LEAST ONCE in 24 hours
p_overwhelmed_24_hours = 1 - p_safe_24_hours

print(f"Probability of being overwhelmed at least once in 24 hours: {p_overwhelmed_24_hours:.4f}")
print(f"That's about {p_overwhelmed_24_hours*100:.2f}%")
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

Exponential distribution

```python
from scipy.stats import expon
# Parameters
mean_arrival = 10  # average time between arrivals (in minutes)

# P(next customer arrives within 5 minutes)
probability = expon.cdf(5, scale=mean_arrival)

print(f"Probability of next customer arriving within 5 minutes: {probability:.4f}")
print(f"That's about {probability*100:.2f}%")
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
# P(no customer for 15 minutes) = P(next arrival takes MORE than 15 minutes)
probability_break = expon.sf(15, scale=mean_arrival)

print(f"Probability of an employee taking a break: {probability_break:.4f}")
print(f"That's about {probability_break*100:.2f}%")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

Normal distribution

```python
from scipy.stats import norm

# Parameters
mu = 150     # average weight
sigma = 10   # standard deviation

# P(140 <= weight <= 160)
probability = norm.cdf(160, mu, sigma) - norm.cdf(140, mu, sigma)

print(f"Probability of weight between 140 and 160 grams: {probability:.4f}")
print(f"That's about {probability*100:.2f}%")
```
